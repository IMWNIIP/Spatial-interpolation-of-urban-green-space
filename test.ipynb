```python
import torch
import random
from torch.autograd import Variable
import matplotlib.pyplot as plt
batchSize = 1
imageSize = 32
nc = 1

```

```python
def ControlPointsImage_random(dems, ncp,):  # Random sampling
    y_cpLayer = torch.FloatTensor(batchSize ,nc, imageSize, imageSize).zero_()
    y_cpLayer = Variable(y_cpLayer)
    cp = []
    for n in range(0, batchSize):
        cp.append([])
        for ite in range(0, ncp):
            i = random.randint(0, imageSize - 1)
            j = random.randint(0, imageSize - 1)
            cp[n].append([dems[n, 0, i, j], i, j])  # extract dem control point function
    for i in range(0, batchSize):
        for _cp in cp[i]:
            h = _cp[0]
            x = _cp[1]
            y = _cp[2]
            y_cpLayer[i, 0, x, y] = h
    return y_cpLayer
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# 读取TIF图片
img = Image.open('D:/cdcgan/data/ph/64xs/64/180.tif')
img_array = np.array(img)

# 获取图片的尺寸
height, width = img_array.shape

# 随机选取100个像素
selected_pixels = np.random.choice(height * width, size=100, replace=False)

# 将选中的像素提取出来
selected_pixels_array = np.zeros((height, width))
for i in selected_pixels:
    y = i // width
    x = i % width
    selected_pixels_array[y, x] = img_array[y, x]

# 使用matplotlib显示图片
#plt.imshow(img_array, cmap='gray')
plt.imshow(selected_pixels_array, cmap='Blues')
plt.show()
```
